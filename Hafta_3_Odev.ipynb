BAZI NOTLAR:

  Eğer veri seti içerisinde yapılan **quartile** sonuçları tam sayı çıkmaz ise **interpolasyon** yapılmaladır.
  **Quartile Formulü:** ( (n + 1) * 0.25 / n=dataset içerisindeki toplam veri sayısı
  **İnterpolasyon Formülü:** {1. değer} + ({2. değer} - {1. değer}) * 0.5) / Arada kalınan değerin sağındaki veri 2. değer, solundaki veri ise 1. değer olarak alınır.

  **Min-Max Ölçekleme:** Dağılım önemsiz, tüm değerleri 0-1 arasına sıkıştırır.
  **Z-Score Standartlaştırması:** Dağılım normal olsa iyi olur, ortalaması 0 standart sapması 1 olacak şekilde sıkıştırır
  **Robust Standartlaştırması:** Outlier sayısı fazla ise veride tercih edilir
  **Logaritmik Dönüşüm:** Veri setinde asimetrik bir dağılım var ise, değerler arasında çok fazla fark var ise eğer

  **Nominal:** Sırasız kategorik veriler, Örneğin: renkler
  **Ordinal:** Sıralı kategorik veriler, Örneğin: ilkokul,ortaokul,lise

  **Label Encoding:** Veri seti içerisindeki etiketler 0'dan başlayarak sırayla değerler alır. Aldıkları değerler sabittir ve ilerde dataset içerisinde yine denk gelirse yine aynı değer gösterilir.
  **One-Hot Encoding:** Her kategori belirli değerler 1 olacak şekilde matrixlerden oluşur. Aldıkları matrix karşılıkları sabittir ve ilerde dataset içerisinde yine denk gelirse yine aynı değer gösterilir.
  **Binary Encoding:** Önce veriler Label Encoding ile sıra değerleri verilir ardından binary sistemdeki karşılıkları yazılır. Aldıkları değerler sabittir ve ilerde dataset içerisinde yine denk gelirse yine aynı değer gösterilir.
  **Ordinal Encoding:** Label Encoding gibi 0'dan başlayarak sırayla artan değerler veriyor. Aldıkları değerler sabittir ve ilerde dataset içerisinde yine denk gelirse yine aynı değer gösterilir.
  **Frequency Encoding:** Veri setindeki değerlerin görülme sıklığı yani modları ne ise o değerleri alırlar. Aldıkları değerler sabittir ve ilerde dataset içerisinde yine denk gelirse yine aynı değer gösterilir.
  **Target Encoding:** Kategoriler, hedef değişkenle (target) ilişkisine göre sayısal değerlere dönüştürülür.Aldıkları değerler sabittir ve ilerde dataset içerisinde yine denk gelirse yine aynı değer gösterilir.

  DUMMY TUZAĞI: One-Hot Encoding yaparken aslında matrix içerisindeki kategorilerden birine fazladan değer atanıyor. Matrix içerisindeki bir kategori eksik olursa eğer biz onun değerini diğer kategorilere atanan verilerden bulabiliriz. Bunu önlemek önemli.

```python
  # Kategorik sütunların seçilmesi işlemi:
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
  # Numerik sütunların seçilmesi işlemi:
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

print('Kategorik Sütunlar:')
print(categorical_columns)
print('\nNumerik Sütunlar:')
print(numerical_columns)
```

```python
  # Numerik sütunlardaki null değerlerin ortalama ile doldurulması işlemi
def fill_numerical_columns(df, numerical_columns):
    for column in numerical_columns:
      df[column].fillna(df[column].mean(), inplace=True)
    return df

    # Sütunlarda boş veri yoksa o sütun içerisinde herhangi bir değişiklik yapmayacak
```

```python
df = fill_numerical_columns(df, numerical_columns)
```

```python
# Sadece numeric sütunlardaki boş değerleri ve veri tiplerini kontrol etmek istiyorum:
"""
for i in numerical_columns:
    print(df[i].isnull().sum())
    print(df[i].dtype)
  """

# Boş değer sayısını ve veri tiplerini yeni bir DataFrame'e kaydedelim

"""
summary = pd.DataFrame({
    'Boş Değer Sayısı': [df[i].isnull().sum() for i in numerical_columns], # Sütun1
    'Veri Tipi': [df[i].dtype for i in numerical_columns] # Sütun2
}, index=numerical_columns)
"""

def check_nullvalues_and_dtypes_for_numeric_columns(df):
  summary = pd.DataFrame({
      'Boş Değer Sayısı': [df[i].isnull().sum() for i in numerical_columns], # Sütun1
      'Veri Tipi': [df[i].dtype for i in numerical_columns] # Sütun2
  }, index=numerical_columns)
  return summary
```

```python
  # Kategorik değerlerin mod ile doldurulması:
def fill_categorical_columns(df, categorical_columns):
    for column in categorical_columns:
      df[column].fillna(df[column].mode()[0], inplace=True)
    return df
```

```python
df = fill_categorical_columns(df, categorical_columns)
```
