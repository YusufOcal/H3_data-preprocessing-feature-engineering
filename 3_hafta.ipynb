#INTERPOLASYON

Interpolasyon, **bir çeyrek pozisyonunun tam bir değeri ifade etmediği** (örneğin, 1.5, 2.75 gibi kesirli pozisyonlar) durumlarda gerekir. Kesirli pozisyon, verinin tam bir gözlem değeri değil, iki veri arasında bir noktada bulunduğunu gösterir. Bu durumda, interpolasyonla iki değerin arasındaki doğru ara değer hesaplanır.

---

### **Interpolasyonun Gerekip Gerekmediği Durumlar:**

#### **Interpolasyon Gerekmeyen Durumlar:**
- **Çeyrek pozisyonları tam bir sayı olduğunda, interpolasyona gerek yoktur.**
  - Pozisyon, veri setindeki sıralı bir gözlemin tam yerini gösterir.
  - Örneğin:
    - Pozisyon \( 1 \): Bu pozisyon 1. sıradaki değerdir.
    - Pozisyon \( 3 \): Bu pozisyon 3. sıradaki değerdir.

---

#### **Interpolasyon Gereken Durumlar:**
- **Çeyrek pozisyonları tam bir sayı olmadığında (kesirli pozisyon olduğunda), interpolasyona ihtiyaç vardır.**
  - Pozisyon, iki veri noktası arasındaki bir noktayı gösterir.
  - Bu durumda, çeyrek değeri bulmak için iki veri arasındaki oransal mesafe (lineer interpolasyon) hesaplanır.
  - Örneğin:
    - Pozisyon \( 1.5 \): Bu pozisyon 1. ve 2. değerler arasında bir noktadadır.

---

### **Örneğe Uygulama:**

Veri seti: \( [2, 4, 6, 8, 10] \)

#### **1. Q1 Pozisyonu:**
- Pozisyon: \( (n + 1) * 0.25 = (5 + 1) *0.25 = 6 * 0.25 = 1.5 \)
- **Kesirli pozisyon:** \( 1.5 \), 1. ve 2. değerler arasında bir noktadadır.
  - **Interpolasyon Gereklidir:**
  
    Q1 = {1. değer} + ({2. değer} - \{1. değer}) * 0.5)
    Q1 = 2 + ((4 - 2) * 0.5) = 2 + 1 = 3


#### **2. Q3 Pozisyonu:**
- Pozisyon: \( (n + 1) * 0.75 = (5 + 1) *0.75 = 6 * 0.75 = 4.5 \)
- **Kesirli pozisyon:** \( 4.5 \), 4. ve 5. değerler arasında bir noktadadır.
  - **Interpolasyon Gereklidir:**

    Q3 = {4. değer} + ({5. değer} - {4. değer}) * 0.5

    Q3 = 8 + (10 - 8) * 0.5 = 8 + 1 = 9


---

#### **Sonuçlar:**
- Interpolasyon, pozisyonlar kesirli olduğunda gereklidir çünkü bu, çeyrek değerin bir veri gözlemi yerine iki gözlem arasındaki bir noktada olduğunu gösterir.
- Eğer pozisyonlar tam bir sayı olsaydı (örneğin \( 1, 2, 3 \)), doğrudan ilgili değeri seçerdik ve interpolasyona gerek kalmazdı.

# SAYISAL DEĞERLERİN STANDARTLAŞTIRILMASI

Sayısal değişkenlerin standartlaştırılması, özellikle makine öğrenimi ve veri analizi süreçlerinde kullanılan önemli bir tekniktir. Veri setinde bulunan değişkenler genellikle farklı ölçeklerde olabilir. Bu durum, özellikle mesafeye dayalı algoritmalarda (örneğin, K-En Yakın Komşu, K-Ortalamalar gibi) modelin performansını olumsuz etkileyebilir. Standartlaştırma işlemi, tüm değişkenlerin aynı ölçek ve aralıkta yer almasını sağlar.

**Standartlaştırmanın Önemi**
* **Farklı Ölçeklerin Dengelenmesi:** Özellikle mesafeye dayalı algoritmalarda, büyük değerli değişkenler daha küçük ölçekli değişkenlerden daha fazla etkili olabilir.
* **Model Performansı:** Daha dengeli ölçekler, modelin daha doğru sonuçlar üretmesini sağlar.
Değişkenler arasındaki korelasyonun doğru bir şekilde yorumlanabilmesini sağlar.



*  Min-Max Ölçekleme (Normalization):Değerler, belirli bir aralığa (örneğin 0 ile 1 arasında) ölçeklenir.Özellikle veri dağılımı normal değilse kullanılır.
*  Z-Score Standartlaştırması:Veriyi, ortalaması 0 ve standart sapması 1 olan bir dağılıma dönüştürür.
*  Robust Standartlaştırma:Medyan ve çeyrek değerler kullanılarak yapılır.Özellikle aykırı değerlere karşı dayanıklıdır.
*  Logaritmik Dönüşüm:Değerleri logaritmik ölçeğe taşır.Özellikle büyük varyansa sahip verilerde etkili olabilir.






Sayısal değişkenlerin standartlaştırılması için bir yöntem seçerken, veri setinin özellikleri, kullanılacak makine öğrenimi modeli ve analizin amacı dikkate alınmalıdır. Aşağıda hangi durumlarda hangi standartlaştırma yönteminin tercih edilmesi gerektiği detaylı olarak açıklanmıştır:

---

### **1. Min-Max Ölçekleme**
#### **Ne Zaman Kullanılır?**
- **Dağılımın şekli:** Değişkenlerin dağılımının normal olup olmamasına bakılmaksızın kullanılabilir.
- **Değer aralıkları:** Tüm değerlerin belirli bir aralıkta (genellikle [0, 1]) olmasını istediğinizde.
- **Kullanılacak model:** Mesafeye dayalı algoritmalarda (örneğin KNN, K-Means, SVM).
- **Küçük varyans:** Değişkenlerin değerleri küçük bir aralıkta sıkışmışsa.

#### **Avantajları:**
- Basit ve hızlı bir yöntemdir.
- Özellikle özellikler aynı aralığa getirildiği için görselleştirmede etkili olabilir.

#### **Dezavantajları:**
- Aykırı değerlerden etkilenir. Veri setinde outlier varsa yanlış sonuçlara yol açabilir.

---

### **2. Z-Score Standartlaştırması**
#### **Ne Zaman Kullanılır?**
- **Dağılımın şekli:** Veri dağılımı normal veya normale yakın olduğunda tercih edilir.
- **Değerlerin merkezi ve yayılımı:** Verileri ortalaması 0 ve standart sapması 1 olacak şekilde ölçeklendirmek istediğinizde.
- **Kullanılacak model:** Lineer regresyon, lojistik regresyon, PCA gibi normal dağılım varsayımı yapan algoritmalarda.
- **Karşılaştırma:** Farklı birimlerdeki değişkenlerin etkilerini eşitlemek istediğinizde.

#### **Avantajları:**
- Aykırı değerlerin etkisi Min-Max ölçeklemeye göre daha azdır.
- Değişkenlerin merkezi konumunu dengeler.

#### **Dezavantajları:**
- Veri dağılımı normal değilse beklenen performansı vermez.

---

### **3. Robust Standartlaştırma**
#### **Ne Zaman Kullanılır?**
- **Aykırı değerler:** Veri setinde outlier'ların olduğu durumlarda.
- **Dağılımın şekli:** Veri dağılımı normal değilse veya değişkenlerin medyan etrafında yoğunlaşması bekleniyorsa.
- **Kullanılacak model:** Aykırı değerlere duyarlı modeller yerine kararlı sonuçlar beklenen analizlerde.

#### **Avantajları:**
- Aykırı değerlere dayanıklıdır.
- Özellikle aykırı değerlerin aşırı baskın olduğu veri setlerinde etkilidir.

#### **Dezavantajları:**
- Özelliklerin varyansını azaltabilir, bu nedenle tüm değişkenler arasında tam eşitlik sağlamaz.

---

### **4. Logaritmik Dönüşüm**
#### **Ne Zaman Kullanılır?**
- **Dağılımın şekli:** Veri setinde asimetrik bir dağılım varsa (örneğin pozitif çarpıklık).
- **Değerlerin ölçeği:** Değerler arasında çok büyük farklar olduğunda ve yüksek varyans varsa.
- **Kullanılacak model:** Lineer regresyon, lojistik regresyon gibi modellerde daha iyi bir doğrusal ilişki kurmak için.

#### **Avantajları:**
- Dağılımı daha normal hale getirir.
- Büyük değerler ile küçük değerler arasındaki farkı azaltır.

#### **Dezavantajları:**
- Negatif veya sıfır değerler ile çalışmak zordur. Bu durumda küçük bir sabit (\(+1\)) eklenir.
- Değişkenlerin interpretasyonunu zorlaştırabilir.

---

### **5. Seçim Kriterleri Tablosu**

| **Kriter**                      | **Min-Max Ölçekleme**       | **Z-Score Standartlaştırması** | **Robust Standartlaştırma**    | **Logaritmik Dönüşüm**          |
|----------------------------------|-----------------------------|---------------------------------|--------------------------------|---------------------------------|
| **Aykırı Değerler**             | Duyarlı                    | Daha az duyarlı               | Dayanıklı                      | Orta düzeyde etkilenir         |
| **Dağılımın Normal Olması**     | Gerekli değil              | Önerilir                      | Gerekli değil                 | Gerekli değil                  |
| **Mesafeye Dayalı Modeller**    | Önerilir                  | Önerilir                      | Duruma bağlı                  | Duruma bağlı                   |
| **Normalleştirme Aralığı**      | [0, 1]                    | (-∞, ∞)                       | (-∞, ∞)                       | (-∞, ∞)                        |
| **Negatif/Sıfır Değerler**      | Çalışabilir               | Çalışabilir                   | Çalışabilir                   | Ek sabit gerekebilir           |
| **Varyans Yüksekliği**          | Orta düzeyde etkili        | İyi                           | İyi                           | Çok etkili                     |

---

### **Sonuç**
1. **Min-Max Ölçekleme:** Modelin çıktısının belli bir aralıkta olması gereken durumlarda.
2. **Z-Score Standartlaştırması:** Verilerin normal dağılım varsayımına uygun olduğu ve modelin bunu beklediği durumlarda.
3. **Robust Standartlaştırma:** Aykırı değerlere duyarlı bir veri setiyle çalışıyorsanız.
4. **Logaritmik Dönüşüm:** Veri dağılımı asimetrik ve varyans büyükse.

## NORMAL DAĞILIM- ÇARPIK DAĞILIM

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Veri setinin oluşturulması
# Çarpık bir dağılım (pozitif çarpıklık için exponansiyel dağılım)
np.random.seed(42)  # Rastgeleliği sabitlemek için
data_skewed = np.random.exponential(scale=2, size=1000)  # Pozitif çarpık dağılım
data_with_outliers = np.append(data_skewed, [50, 60, 70])  # Aykırı değerler ekleniyor

# Veri seti DataFrame'e dönüştürülüyor
df = pd.DataFrame(data_with_outliers, columns=["Skewed_Data"])
print("Normal Dağılım Göstermeyen Veri Seti (ilk 10 satır):")
print(df.head(10))

# Histogram ile görselleştirme
plt.hist(data_with_outliers, bins=30, alpha=0.7, color="red", edgecolor="black")
plt.title("Normal Dağılım Göstermeyen Histogram")
plt.xlabel("Değer")
plt.ylabel("Frekans")
plt.show()

````

````python
from scipy.stats import shapiro

# Shapiro-Wilk testi
stat, p = shapiro(data_with_outliers)
print("Shapiro-Wilk Testi:")
print(f"Test İstatistiği = {stat}, P-değeri = {p}")
if p > 0.05:
    print("Veri normal dağılım göstermektedir.")
else:
    print("Veri normal dağılım göstermemektedir.")
````

````python
import numpy as np
import pandas as pd

# Normal dağılım parametreleri
ortalama = 50  # Ortalama (mean)
standart_sapma = 10  # Standart sapma (std)
veri_sayisi = 1000  # Veri sayısı

# Normal dağılım verisi oluşturma
normal_veri = np.random.normal(loc=ortalama, scale=standart_sapma, size=veri_sayisi)

# Veri seti oluşturma
df = pd.DataFrame(normal_veri, columns=["Normal_Dağılım"])
print("Normal Dağılım Gösteren Veri Seti (ilk 10 satır):")
print(df.head(10))

# Histogram ile dağılımı görselleştirme
import matplotlib.pyplot as plt

plt.hist(normal_veri, bins=30, alpha=0.7, color="blue", edgecolor="black")
plt.title("Normal Dağılım Histogramı")
plt.xlabel("Değer")
plt.ylabel("Frekans")
plt.show()
````

````python
from scipy.stats import shapiro

stat, p = shapiro(normal_veri)
print("Shapiro-Wilk Testi:")
print(f"Test İstatistiği = {stat}, P-değeri = {p}")
if p > 0.05:
    print("Veri normal dağılım göstermektedir.")
else:
    print("Veri normal dağılım göstermemektedir.")
````

# Kategorik Verilerin Etiketlenmesi (Encoding)
Makine öğrenmesi ve veri analizi süreçlerinde çok önemli bir adımdır. Bu işlem, modelin verileri anlaması ve işleyebilmesi için kategorik (metinsel veya sınıfsal) verilerin sayısal bir formata dönüştürülmesini içerir.
---

### 1. **Kategorik Veri Nedir?**
Kategorik veri, belirli sınıflar veya kategoriler içeren verilerdir.
- **Nominal Veri:** Sıralama olmayan kategoriler (örnek: renkler - kırmızı, mavi, yeşil).
- **Ordinal Veri:** Sıralama içeren kategoriler (örnek: eğitim düzeyi - ilkokul, lise, üniversite).

### 2. **Kategorik Verilerin Etiketlenme Yöntemleri**

#### a) **Label Encoding (Etiketleme Yöntemi)**
- Kategoriler sayısal değerlerle temsil edilir. Her kategoriye bir tamsayı atanır.
- Örnek:
  - Kategoriler: ["Kırmızı", "Mavi", "Yeşil"]
  - Label Encoding: {"Kırmızı": 0, "Mavi": 1, "Yeşil": 2}
- **Avantajları:**
  - Basit ve hızlıdır.
  - Modelin sıralı verilerdeki ilişkiyi öğrenmesine izin verir.
- **Dezavantajları:**
  - Sırasız verilerde yanlış anlamalara neden olabilir (örneğin, model "Kırmızı" < "Mavi" < "Yeşil" ilişkisini öğrenebilir, bu da hatalıdır).

#### b) **One-Hot Encoding**
- Her kategori, ikili (binary) bir vektörle temsil edilir.
- Örnek:
  - Kategoriler: ["Kırmızı", "Mavi", "Yeşil"]
  - One-Hot Encoding:
    - Kırmızı: [1, 0, 0]
    - Mavi: [0, 1, 0]
    - Yeşil: [0, 0, 1]
- **Avantajları:**
  - Kategoriler arasında herhangi bir sıralama olmadığından modelin yanlış ilişkiler öğrenmesi önlenir.
- **Dezavantajları:**
  - Kategori sayısı çok olduğunda (örneğin, 1000'den fazla kategori) hafıza tüketimi ve hesaplama maliyeti artar.

#### c) **Binary Encoding**
- İlk olarak, kategoriler label encoding ile sayısal değerlere dönüştürülür.
- Daha sonra bu sayısal değerler ikili (binary) formatta temsil edilir.
- Örnek:
  - Kategoriler: ["Kırmızı", "Mavi", "Yeşil"]
  - Label Encoding: {"Kırmızı": 0, "Mavi": 1, "Yeşil": 2}
  - Binary Encoding:
    - Kırmızı: [0]
    - Mavi: [1]
    - Yeşil: [10]
- **Avantajları:**
  - One-Hot Encoding'e göre daha az bellek kullanır.
- **Dezavantajları:**
  - Özellik mühendisliği sürecinde dikkatli olunmalıdır.

#### d) **Ordinal Encoding**
- Sıralı veriler için uygundur. Kategorilere sırasıyla artan tamsayılar atanır.
- Örnek:
  - Kategoriler: ["İlkokul", "Lise", "Üniversite"]
  - Ordinal Encoding: {"İlkokul": 0, "Lise": 1, "Üniversite": 2}
- **Avantajları:**
  - Sıralı veriler için anlamlıdır.
- **Dezavantajları:**
  - Sırasız kategorilerde hatalı ilişki öğrenilmesine neden olabilir.

#### e) **Frequency Encoding**
- Her kategori, veri setindeki görülme sıklığı ile temsil edilir.
- Örnek:
  - Kategoriler: ["Kırmızı", "Mavi", "Kırmızı", "Yeşil", "Mavi"]
  - Frequency Encoding: {"Kırmızı": 2, "Mavi": 2, "Yeşil": 1}
- **Avantajları:**
  - Kategori sıklığını hesaba katar.
- **Dezavantajları:**
  - Modelin verileri yanlış yorumlamasına neden olabilir (örneğin, sıklık ilişkisi öğrenmesi gerekmediği durumlarda).

#### f) **Target Encoding (Hedef Kodlama)**
- Kategoriler, hedef değişkenle (target) ilişkisine göre sayısal değerlere dönüştürülür.
- Örnek:
  - Kategoriler: ["A", "B", "C"]
  - Hedef Değişken Ortalaması: {"A": 0.6, "B": 0.3, "C": 0.8}
- **Avantajları:**
  - Kategoriler ile hedef değişken arasındaki ilişkileri modele taşıyabilir.
- **Dezavantajları:**
  - Aşırı öğrenme (overfitting) riski taşır.

---

### 3. **Hangi Yöntemi Seçmeliyim?**
- **Veri Seti Boyutu ve Kategori Sayısı:**
  - Az sayıda kategori için **One-Hot Encoding**.
  - Çok sayıda kategori için **Binary Encoding** veya **Frequency Encoding**.
- **Veri Türü:**
  - Sıralı kategoriler için **Ordinal Encoding**.
  - Sırasız kategoriler için **One-Hot Encoding**.
- **Hedef Değişken ile İlişki:**
  - Hedef değişkenle güçlü bir ilişki varsa **Target Encoding**.

- **Label Encoding**, model türüne bağlı olarak kullanılabilir.
Label Encoding, sırasız kategorik verilerde Linear Regression, SVM gibi algoritmalar kullanılıyorsa dikkatli olunmalıdır. Çünkü model bu değerler arasında bir büyüklük ilişkisi olduğunu varsayabilir ve bu yanlış sonuçlara yol açabilir.

````python
import matplotlib.pyplot as plt
from scipy.stats import shapiro, normaltest, probplot

for column in numeric_columns:
    plt.hist(data[column], bins=30, alpha=0.7, color="blue", edgecolor="black")
    plt.title(f"{column.capitalize()} Dağılımı Histogramı")
    plt.xlabel(column.capitalize())
    plt.ylabel("Frekans")
    plt.show()

    # Shapiro-Wilk Testi
    stat, p = shapiro(data[column])
    print(f"Shapiro-Wilk Testi: Test İstatistiği = {stat:.4f}, p-değeri = {p:.4f}")

    # Normality Test (D'Agostino's K^2 Testi)
    stat, p = normaltest(data[column])
    print(f"D'Agostino's K^2 Testi: Test İstatistiği = {stat:.4f}, p-değeri = {p:.4f}")

    # Karar
    if p > 0.05:
        print("Sonuç: Veri normal dağılıma uygun görünüyor.")
    else:
        print("Sonuç: Veri normal dağılıma uygun değil.")
````

````python
for col in numeric_columns:
    print(f"\n--- {col} Sütunu ---")

    # Eksik değer kontrolü ve doldurma
    if data[col].isnull().sum() > 0:
        print(f"{col} sütununda {data[col].isnull().sum()} eksik değer var. Dolduruluyor...")
        data[col].fillna(data[col].mean(), inplace=True)  # Eksik değerleri sütunun ortalaması ile doldur

    # Boxplot ile aykırı değerleri görselleştir
    plt.boxplot(data[col], vert=False, patch_artist=True)
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()

    # IQR yöntemi ile aykırı değer analizi
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(f"IQR: {IQR}, Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    print(f"Aykırı Değer Sayısı: {len(outliers)}")


    # Aykırı değerleri sınıra çek
    data[col] = data[col].apply(
        lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x
    )

    # Sınıra çekilmiş veriyi tekrar görselleştir
    plt.boxplot(data[col].dropna(), vert=False, patch_artist=True)
    plt.title(f"Boxplot of {col} (Sınıra Çekilmiş)")
    plt.xlabel(col)
    plt.show()
````
